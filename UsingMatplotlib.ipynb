```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Create a NumPy array x containing 100 evenly spaced values between 0 and 2.
x = np.linspace(0, 2 * np.pi, 100)
#Calculate y1=sin(x) and y2=cos(x).
y1 = np.sin(x)
y2 = np.cos(x)

#Create a single figure and axes using plt.figure() and plt.plot().
plt.figure(figsize=(8, 5))
#Plot y1 vs. x and y2 vs. x on the same axes.
plt.plot(x, y1, label='Sine', color='blue', linestyle='-')
plt.plot(x, y2, label='Cosine', color='orange', linestyle='--')

#Add a title to the plot: "Sine and Cosine Waves".
#plt.title("Sine and Cosine Waves")

#Label the x-axis: "Angle (radians)".
plt.xlabel("Angle (radians)")

#Label the y-axis: "Value".
plt.ylabel("Value")

#Add a legend to distinguish the two lines (label y1 as "Sine" and y2 as "Cosine").
plt.legend()

#Display the plot using plt.show().
plt.show()

#Scatter Plot and Customization
#Create 50 random data points for temp (e.g., between 15 and 35) using np.random.uniform().
np.random.seed(42) # for reproducibility
temp = np.random.uniform(15, 35, 50)

#Create 50 corresponding data points for sales that show a positive correlation with temp, plus some random noise (e.g., sales = 5 * temp + np.random.normal(0, 10, 50)).
sales = 5 * temp + np.random.normal(0, 10, 50)

#Create a scatter plot with temp on the x-axis and sales on the y-axis using plt.scatter().
plt.figure(figsize=(7, 6))
#Set the color of the points to 'red'.
#Set the marker shape to a triangle ('^').
#Set the size of the markers to 50.
#Add a grid to the plot.
plt.scatter(temp,
            sales,
            color='red',        # Set color
            marker='^',         # Set marker shape
            s=50,               # Set marker size
            alpha=0.7,          # Optional: Add transparency
            label='Data Points')
plt.show()

#Bar Chart and Subplots (Categorical Data)
#Define a list of categories: months = ['Jan', 'Feb', 'Mar', 'Apr'].
months = ['Jan', 'Feb', 'Mar', 'Apr']

#Define corresponding values: revenue = [150, 210, 180, 300].
revenue = [150, 210, 180, 300]

#Create a figure that contains two subplots arranged vertically (2 rows, 1 column).
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 8))

#In the top subplot, create a bar chart of revenue vs. months.
axes[0].bar(months, revenue, color='skyblue')
#Set the title of this subplot to "Monthly Revenue".
axes[0].set_title("Monthly Revenue (Vertical View)")
axes[0].set_ylabel("Revenue (Thousands)")


#In the bottom subplot, create a horizontal bar chart using plt.barh().
#Color the bars in this subplot 'green'.
axes[1].barh(months, revenue, color='green')
#Set the title of this subplot to "Revenue (Horizontal View)".
axes[1].set_title("Monthly Revenue (Horizontal View)")
axes[1].set_xlabel("Revenue (Thousands)")


#Use plt.tight_layout() to prevent titles and labels from overlapping.
plt.tight_layout()

plt.show()


```
